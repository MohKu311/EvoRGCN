```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define file names
mode_files = [
    "df_actions_27k_reaction.csv",
    "df_actions_27k_activation.csv",
    "df_actions_27k_catalysis.csv",
    "df_actions_27k_binding.csv",
    "df_actions_27k_inhibition.csv",
    "df_actions_27k_ptmod.csv",
    "df_actions_27k_expression.csv",
]
```

```python
# Load each CSV file into a dictionary of DataFrames
df_action_splits = {file.split("_")[-1].split(".")[0]: pd.read_csv(file) for file in mode_files}
```

```python
# Display the first few rows of each DataFrame
for mode, df_mode in df_action_splits.items():
    print(f"Displaying dataframe for mode: {mode}")
    display(df_mode.head())
```

```python
# Create a weighted NetworkX graph for each mode
graphs = {}
```

```python
for mode, df_mode in df_action_splits.items():
    G = nx.Graph()
    edges = list(zip(df_mode["item_id_a"], df_mode["item_id_b"], df_mode["score"]))
    G.add_weighted_edges_from(edges)
    graphs[mode] = G
    print(f"Created Weighted NetworkX graph for {mode}")
```

```python
# Plot weighted NetworkX graphs for each mode
for mode, G in graphs.items():
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G, seed=42)  # Layout for visualization
    nx.draw(G, pos, node_size=10, edge_color="gray", alpha=0.6, with_labels=False)
    edge_labels = {(u, v): f'{d:.1f}' for u, v, d in G.edges(data="weight")}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)
    plt.title(f"Weighted Network Graph for {mode}")
    plt.show()
```
